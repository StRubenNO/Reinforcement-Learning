The above figure is what you will try to approximate through the TD prediction algorithm.

Your algorithm for TD prediction has five arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `policy`: This is a 1D numpy array with `policy.shape` equal to the number of states (`env.nS`).  `policy[s]` returns the action that the agent chooses when in state `s`.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `V`: This is a dictionary where `V[s]` is the estimated value of state `s`.

Please complete the function in the code cell below.

````python
from collections import defaultdict, deque
import sys

def td_prediction(env, num_episodes, policy, alpha, gamma=1.0):
    # initialize empty dictionaries of floats
    V = defaultdict(float)
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODO: complete the function
        # Implement Algorithm 12 in
        # https://github.com/udacity/rl-cheatsheet/blob/master/cheatsheet.pdf
        
        # begin an episode, observe S_0
        # Start at state 36, for this specific environment
        s_0 = env.reset()
        state = s_0
        
        # Loop through all states
        t=0
        while True:
            # Take action based on the state:
            
            # Recommended action:
            A_state = policy[state]
            
            #Based on recommended action:
            next_state, reward, terminal_state, probality = env.step(A_state)
            #print(env.step(A_state))
            #print(terminal_state)
            
            V[state] = V[state] + alpha * (reward + gamma * V[next_state] - V[state])
            
            state = next_state
            
            t += 1
            
            if terminal_state:
                  break
        
    return V 
````

### Part 2: TD Control: Sarsa

In this section, you will write your own implementation of the Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def update_Q(Qsa, Qsa_next, reward, alpha, gamma):
    """ updates the action-value function estimate using the most recent time step """
    return Qsa + (alpha * (reward + (gamma * Qsa_next) - Qsa))

def policy_epsilon_greedy(env, Q_state, epsilon=0.5):
   
    # Define policy with equal an low possibility for each action
    policy_epsilon = np.ones(env.nA) * epsilon / env.nA
    
    # Greedy policy: https://classroom.udacity.com/nanodegrees/nd009-ent/parts/80594ce0-6c97-4c9b-9158-08e626ee2215/modules/54c65a40-fd5d-4715-883e-22e5afa91140/lessons/a653f5f4-9618-4f52-9b2c-f8187883a8b8/concepts/945c310b-493e-4d9c-8f16-1f3d3b892cbb
    
    # Add posibility to the action that is the most greedy
    policy_epsilon[np.argmax(Q_state)] = 1 - epsilon/env.nA*(env.nA-1)
      
    # Select action from a row of actions, with emphasis on the highest p in policy_epsilon
    action = np.random.choice(np.arange(env.nA), p=policy_epsilon)
    
    #print(epsilon, '   ', Q_state, '   ', policy_epsilon, '   ', action)

    return action, policy_epsilon
````

````python
import matplotlib.pyplot as plt
%matplotlib inline


def sarsa(env, num_episodes, alpha, gamma=1.0):
    
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    
    # initialize performance monitor
    plot_every = 100
    tmp_scores = deque(maxlen=plot_every)
    scores = deque(maxlen=num_episodes)
    
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()   
        
        ## TODO: complete the function
        # Implement Algorithm 13 in
        # https://github.com/udacity/rl-cheatsheet/blob/master/cheatsheet.pdf
           
        # For monitoring: initialize score
        score = 0

        # Adjust epsilon for each episode
        epsilon = 1/i_episode
        
        # Prevent epsilon to be close to 0, wich is max greedy
        epsilon_maxGreedy = 0.01
        if epsilon < epsilon_maxGreedy: epsilon=epsilon_maxGreedy
            
        # begin an episode, observe S_0
        # Start at state 36, for this specific environment
        s_0 = env.reset()
        state = s_0
        
        #print(Q[state])
        # Choose action A0 using policy derived from Q (e.g. Epsilon-greedy)
        a0, _ = policy_epsilon_greedy(env, Q[state], epsilon)
        action = a0
        
        #print(i_episode, '   ', epsilon, '   ', Q[state], '   ', action)
        
        # Loop through all states. t_max should be about 300
        t=0
        while True:
            # Take action based on the current state of the environment (env.step):
            #print(env.step(action))
            next_state, reward, terminal_state, probality = env.step(action)

            # Choose action using policy derived from Q (e.g. Epsilon-greedy)
            action_epsilon, _ = policy_epsilon_greedy(env, Q[next_state], epsilon)

            # update_Q(Qsa, Qsa_next, reward, alpha, gamma):
            Q[state][action] = update_Q(Q[state][action], \
                                        Q[next_state][action_epsilon], \
                                        reward, \
                                        alpha, \
                                        gamma)            
            
            state = next_state
            action = action_epsilon
            
            t += 1
            
            # For monitoring: add reward to score
            score += reward
            
            if terminal_state or t> 300:
                tmp_scores.append(score)
                break
                
        if (i_episode % plot_every == 0):
            scores.append(np.mean(tmp_scores))
    
    # plot performance
    plt.plot(np.linspace(0,num_episodes,len(scores),endpoint=False),np.asarray(scores))
    plt.xlabel('Episode Number')
    plt.ylabel('Average Reward (Over Next %d Episodes)' % plot_every)
    plt.show()
    # print best 100-episode performance
    print(('Best Average Reward over %d Episodes: ' % plot_every), np.max(scores))
    
    
    return Q
````
